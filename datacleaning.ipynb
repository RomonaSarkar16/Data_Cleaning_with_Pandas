2) Remove all rows with NULL values- TO replace the NULL values in rows we need to change the original dataframe by using inplace = True 
It modifies the original DataFrame directly instead of creating a new one.
If inplace=False (default), the function returns a new modified DataFrame without changing the original one.
________________________________________
Key Differences
Feature	      dropna()                 (without inplace=True)	         dropna(inplace=True)
Returns      new DataFrame?	              ✅ Yes                               	❌ No
Modifies original DataFrame?	           ❌ No	                                ✅ Yes
Memory efficient?	                     ❌ No (creates a copy)	         ✅ Yes (modifies in place)
________________________________________
 
                                    When to Use What?
Use dropna() without inplace=True when you want to keep the original DataFrame unchanged and assign the cleaned version to a new variable.
Use dropna(inplace=True) when you want to clean the DataFrame without creating a new copy.


### **1. Mean (Average)**
- The mean is the sum of all values divided by the total number of values.
- It is **sensitive to outliers** (extremely high or low values).

**Formula:**
\[
\text{Mean} = \frac{\sum X}{N}
\]
Where:
- \( X \) = Sum of all values
- \( N \) = Number of values


### **2. Median (Middle Value)**
- The median is the middle value when the data is sorted in ascending order.
- If there are **odd** numbers of values, it's the middle one.
- If there are **even** numbers of values, it's the average of the two middle values.
- **Less affected by outliers** than the mean.

### **3. Mode (Most Frequent Value)**
- The **mode** is the value that appears **most frequently** in a dataset.
- There can be **one mode (unimodal), multiple modes (multimodal), or no mode** if all values appear the same number of times.


**Key Differences**
| Measure  | Definition | Affected by Outliers? | Used For |
|----------|------------|----------------------|-----------|
| **Mean** | Average of all values | Yes | Normal distributions |
| **Median** | Middle value | No | Skewed distributions |
| **Mode** | Most frequent value | No | Categorical data |

### **When to Use What?**
- Use Mean when there are **no extreme outliers**.
- Use Median when data has **outliers** or is **skewed**.
- Use Mode when dealing with **categorical** data or finding the **most common value**.


```python
#Calculate the MODE, and replace any empty values with it
import pandas as pd
dataframe_3 = pd.read_csv(r"C:\Users\Romona Sarker\Downloads\data.csv")
x = dataframe_3["Calories"].mode()[0]
# dataframe_3["Calories"].fillna(x,inplace = True)
dataframe_3.fillna({"Calories": x}, inplace = True)
print(dataframe_3.to_string())
```

```python
#To Drop a row which has a NULL value in the "Duration" column
dataframe_3.dropna(subset= ["Duration"], inplace = True)
print(dataframe_3.to_string())
```

Removing Rows
Another way of handling wrong data is to remove the rows that contains wrong data.

This way you do not have to find out what to replace them with, and there is a good chance you do not need them to do your analyses.

```python
# "NaN" should be checked using pd.isna() or pd.isnull(), not as a string.
# dropna() is used to drop rows or columns with missing values, not specific rows by index.
for x in df1.index:
    if pd.isna(df1.loc[x, "Salary"]):
        df1.drop(x, inplace=True)

df1.reset_index(drop=True, inplace=True)  # Reset index after dropping rows
print(df1.to_string())
```
